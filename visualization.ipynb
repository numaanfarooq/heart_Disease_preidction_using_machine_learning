```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
plt.rcParams['font.size'] = 12
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
df = pd.read_csv('dataset.csv')
```

```python

print("Dataset Shape:", df.shape)
print("\nFirst 5 Rows:")
print(df.head())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

```python
#Target value distribution
plt.figure(figsize=(8,5))
sns.countplot(x='target', data=df)
plt.title('Distribution of Heart Disease Cases')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Count')
```

```python
#Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='target', bins=30, kde=True)
plt.title('Age Distribution by Heart Disease Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,14))
sns.histplot(x='sex', data=df, hue='target')
plt.title('Sex Distribution by Heart Disease Status')
plt.xlabel('Sex- 0(female), 1(male)')
plt.ylabel('Count')
```

```python
# Set style for visualizations
sns.set_style('whitegrid')
plt.rcParams['font.size'] = 12
plt.rcParams['figure.figsize'] = (12, 6)

# Select numerical features
numerical_features = df.select_dtypes(include=['int', 'float']).columns.tolist()
numerical_features.remove('target')  

# Plot numerical features vs target
for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    
    # Use a violin plot
    sns.violinplot(x='target', y=feature, data=df, palette='coolwarm', split=True)
    
    # Add titles and labels
    plt.title(f'{feature} Distribution by Heart Disease Status', fontsize=16)
    plt.xlabel('Heart Disease (0 = No, 1 = Yes)', fontsize=14)
    plt.ylabel(feature, fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Add mean lines for better interpretation
    means = df.groupby('target')[feature].mean()
    for i, mean in enumerate(means):
        plt.hlines(mean, i - 0.5, i + 0.5, colors='black', linestyles='dashed', label=f'Mean: {mean:.2f}')
    
    # Add legend
    plt.legend(title='Mean Values', loc='upper right')
    
    plt.show()
```

```python

# Correlation matrix
plt.figure(figsize=(12, 8))
corr_matrix = df.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()

```
